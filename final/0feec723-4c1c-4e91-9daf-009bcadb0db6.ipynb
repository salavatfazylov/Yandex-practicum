## Решение задач 

### Cколько книг вышло после 1 января 2000 года

```python
query = '''
           SELECT COUNT(book_id)
           FROM books
           WHERE publication_date >= '2000-01-01'
        '''

count_book = pd.io.sql.read_sql(query, con = engine)
count_book
```

### Количество обзоров и средняя оценка для каждой книги  

```python
query = '''
           WITH 
           r AS (
               SELECT book_id, count(review_id)
               FROM reviews
               GROUP BY 1
               ),
           b AS (
                SELECT *
                FROM books
                ),
           i AS (
                SELECT book_id, AVG(rating)
                FROM ratings
                GROUP BY 1)
            
            SELECT b.book_id, b.title, r.count, i.avg
            FROM b
            LEFT JOIN r ON b.book_id = r.book_id
            LEFT JOIN i ON b.book_id = i.book_id
            
            
            
            
        '''

count_rev = pd.io.sql.read_sql(query, con = engine)
count_rev.head()
```

### Издательство, которое выпустило наибольшее число книг толще 50 страниц 

```python
query = '''         
            
            SELECT publisher, COUNT(title)
           FROM books as b
           LEFT JOIN publishers AS pu ON b.publisher_id  = pu.publisher_id
           WHERE num_pages > 50 
           GROUP BY publisher
           ORDER BY count DESC
           LIMIT 1
                
            
        '''

fat = pd.io.sql.read_sql(query, con = engine)
fat.head()
```

*Больше всего книг толще 50 страниц выпустило издательство Penguin Books, выпустившее 42 книги*

### Автор с самой высокой средней оценкой книг (только книги с 50 и более оценками)

```python
query = '''         
            SELECT author, AVG(t.avg)
            FROM (SELECT title, COUNT(rating), AVG(rating)
                FROM books as b
                LEFT JOIN ratings AS r ON b.book_id  = r.book_id
                GROUP BY title) AS t
                LEFT JOIN books AS b ON t.title=b.title
                LEFT JOIN authors AS a ON b.author_id=a.author_id
                WHERE t.count >= 50
                GROUP BY author
                ORDER BY avg DESC
                LIMIT 1
                   
            
        '''

top = pd.io.sql.read_sql(query, con = engine)
top.head()
```

*Автором с самой высокой средней оценкой с более 50 оценками является J.K. Rowling/Mary GrandPré, у него рейтинг 4.283844*

### Среднее количество обзоров от пользователей, которые поставили больше 50 оценок.

```python
query = '''         
            
             
            SELECT AVG(count)
            FROM (
                SELECT r.username, COUNT(rating)
                FROM (
                    SELECT r.username, COUNT(rating_id) as rating
                    FROM books AS b
                    LEFT JOIN ratings AS r ON b.book_id  = r.book_id
                    GROUP BY r.username
                ) AS u
                RIGHT JOIN reviews AS r ON u.username  = r.username
                WHERE rating > 50
                GROUP BY r.username
                ) AS t
            
            
            
                           
            
        '''

top = pd.io.sql.read_sql(query, con = engine)
top.head()
```

*Среднее количество обзоров от пользователей с более 50 оценками 24,33*
